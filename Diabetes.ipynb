```python
import pandas as pd
data=pd.read_csv("G:\\MY_PLACE\\machine learning\\Diabetes\\Diabetes\\diabetes.csv")
```

```python
data.head()
```

```python
#Transformation
data.transform(lambda x:x-x.mean())
```

```python
data.groupby('Outcome')[['BMI']].mean()
```

```python
data.groupby(['Outcome','Pregnancies'])['BMI'].aggregate('mean').unstack()
```

```python
data.pivot_table('BMI',index='Outcome',columns='Pregnancies')
```

```python
data.pivot_table('BMI',['Outcome','BloodPressure'],columns='Pregnancies')
```

```python
#At times itâ€™s useful to compute totals along each grouping. This can be done via the margins keyword
```

```python
data.pivot_table('BMI',['Outcome','BloodPressure'],columns='Pregnancies',margins=True)
```

```python
#aggFuncs to find mean or median etc
data.pivot_table(index='BMI',columns='Pregnancies',aggfunc={'Glucose':'mean'})
```
