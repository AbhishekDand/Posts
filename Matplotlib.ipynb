# Matplotlib 
It was conceived by John Hunter in
2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting
via gnuplot from the IPython command line. IPython’s creator, Fernando Perez, was
at the time scrambling to finish his PhD, and let John know he wouldn’t have time to
review the patch for several months. John took this as a cue to set out on his own, and
the Matplotlib package was born, with version 0.1 released in 2003.

# To create visualization

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

To plot basic equation graphs

```python
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.tanh(x),'*')
plt.plot(x, np.cos(x),'--')
```

# Saving the figs

```python
x = np.linspace(0, 10, 100)
fig=plt.figure()
plt.plot(x, np.sin(x),'*')
plt.plot(x, np.cos(x),'--')
fig.savefig('my_figure.png')
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

# Decorating plots in Matplotlib

```python
plt.plot(x,np.sin(x))
plt.text(0,1,"PANDAS")
plt.plot(x,np.cos(x))
plt.text(3,0,"PYTHON") # adding text in the graph
plt.xlim(-1,5)#setting xlimits
plt.ylim(-1.5,1)# setting ylimits
plt.title("pandaspython")
plt.xlabel("x")
plt.ylabel("output")
plt.legend(["sin_curve","cos_curve"])
```

# Scatter Plot

A scatter plot (also called a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram) is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data.

Let's create our own data and visualize scatterplot

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
color = list(range(0,13,1))
size=list(range(13,0,-1))
plt.scatter(x,y,c=color,s=size)
```

# SubPlots

```python
fig = plt.figure() # to create figure for multiple graphs
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
ylim=(-1.2, 1.2))
x = np.linspace(0, 10)
ax1.plot(np.sin(x)) #graph 1
ax2.plot(np.cos(x)); #graph 2
```

```python
# Subplot for multiple graph plotting
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.plot(5+i)
```

```python
grid = plt.GridSpec(3, 3, wspace=0.2, hspace=0.3)
```

```python
plt.subplot(grid[0, 0:]) #[row,column pair]
plt.subplot(grid[1, 1:])
plt.subplot(grid[2,:2]);
```

# Three Dimensional Plotting

```python
from mpl_toolkits import mplot3d
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**3))

x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='viridis')
```
